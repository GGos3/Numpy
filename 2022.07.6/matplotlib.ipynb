# 기본 그래프

## 기본 그래프 그리기
- plt.plot([X 축], [Y 축])

```python
# 기본 그래프 그리기

import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40])
plt.show()
```

```python
# plt.plot([X 축], [Y 축])

plt.plot([1, 2, 3, 4], [12, 43, 25, 15])
plt.show()
```

## 그래프에 제목 넣기

```python
# 그래프에 제목 넣기

plt.title('plotting')
plt.plot([10, 20, 30, 40])
plt.show()
```

## 그래프에 범례 넣기

```python
# 그래프에 범례 넣기

plt.title('legend')
plt.plot([10, 20, 30, 40], label = 'asc') # 증가를 의미라는 asc 범례
plt.plot([40, 30, 20, 10], label = 'desc') # 감소를 의미하는 desc 범례0
plt.legend()
plt.show
```

## 그래프 색상 바꾸기

```python
# 그래프 색상 바꾸기

plt.title('color') # 제목 설정
# 그래프 그리기
plt.plot([10, 20, 30 , 40], color = "skyblue", label = 'skyblue')
plt.plot([40, 30, 20, 10], 'pink', label = 'pink')
plt.legend() # 범례 표시
plt.show()
```

## 그래프 선 모양 바꾸기

```python
# 그래프 선 모양 바꾸기

plt.title('plotting')
plt.plot([10, 20, 30, 40], color = 'r', linestyle = '--', label = 'dashed') # 증가
plt.plot([40, 30, 20, 10], 'g', ls = ':', label = 'dotted') # 감소
plt.legend()
plt.show()
```

## 그래프 점 찍기

```python
# 그래프 점 찍기

plt.title('marker')
plt.plot([10, 20, 30, 40], 'r.', label = 'circle') # 증가
plt.plot([40, 30, 20 ,10], 'g^', label = 'triangle up') # 감소
plt.legend()
plt.show()
```

## next() 함수를 활용해 헤더 저장하기

```python
import csv

f = open('Seoul_Data.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
header = next(data) # 헤더를 저장해 값만 불러옴
for row in data :
    print(row)
f.close()
```

```python
import csv

f = open('Seoul_Data.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
header = next(data)
for row in data :
    row[-1] = float(row[-1]) # 정수형을 실수형으로 변경 
    print(row)
f.close()
```
