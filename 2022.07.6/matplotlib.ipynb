## 기본 그래프 그리기
- plt.plot([X 축], [Y 축])

```python
# 기본 그래프 그리기

import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40])
plt.show()
```

## 그래프에 제목 넣기

```python
# 그래프에 제목 넣기

plt.title('plotting')
plt.plot([10, 20, 30, 40])
plt.show()
```

## 그래프에 범례 넣기

```python
# 그래프에 범례 넣기

plt.title('legend')
plt.plot([10, 20, 30, 40], label = 'asc') # 증가를 의미라는 asc 범례
plt.plot([40, 30, 20, 10], label = 'desc') # 감소를 의미하는 desc 범례0
plt.legend()
plt.show
```

## 그래프 색상 바꾸기

```python
# 그래프 색상 바꾸기

plt.title('color') # 제목 설정
# 그래프 그리기
plt.plot([10, 20, 30 , 40], color = "skyblue", label = 'skyblue')
plt.plot([40, 30, 20, 10], 'pink', label = 'pink')
plt.legend() # 범례 표시
plt.show()
```

## 그래프 선 모양 바꾸기

```python
# 그래프 선 모양 바꾸기

plt.title('plotting')
plt.plot([10, 20, 30, 40], color = 'r', linestyle = '--', label = 'dashed') # 증가
plt.plot([40, 30, 20, 10], 'g', ls = ':', label = 'dotted') # 감소
plt.legend()
plt.show()
```

## 그래프 점 찍기

```python
# 그래프 점 찍기

plt.title('marker')
plt.plot([10, 20, 30, 40], 'r.', label = 'circle') # 증가
plt.plot([40, 30, 20 ,10], 'g^', label = 'triangle up') # 감소
plt.legend()
plt.show()
```

## 최고 기온 찾기

```python
import csv

f = open('200506-seoul-weather.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
header = next(data)
max_temp = -1
max_data = ""
for row in data :
    row[-1] = float(row[-1]) # 정수형을 실수형으로 변경 
    if max_temp < row[-1]: # 최고 기온 찾기
        max_data = row[1]
        max_temp = row[-1]
f.close()

print('가장 기온이 높았던 날은 ', max_data + '로 온도는 ', max_temp, '이다.')
```

## 데이터를 읽고 시각화

```python
# 기존 코드
import csv
import matplotlib.pyplot as plt

f = open('200506-seoul-weather.csv', encoding='cp949')
data =csv.reader(f)
header = next(data)
result = []

for row in data:
    if row[3] != '':
        result.append(float(row[3]))
f.close()

plt.plot(result, 'r')
plt.show()
```

# 우리동네 인구 구조 시각화하기

```python
import matplotlib.pyplot as plt
import csv

f = open('age.csv', 'r', encoding='cp949')
data = csv.reader(f)
result = []

for row in data:
    if '양평제2동' in row[0]:
        for i in row[3:]:
            result.append(int(i))

plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

# 대중교통 데이터 시각화 하기

```python
import matplotlib.pyplot
import csv

f = open('subwayfee.csv', encoding='cp949')
data = csv.reader(f)
result = []
next(data)
mx = 0
rate = 0
mx_station = ""

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i])
    if row[6] != 0 and (row[4] + row[6] > 100000):
        rate = row[4] / (row[4] + row[6])
        if rate > mx:
            mx = rate
            mx_station = row[3] + ' ' + row[1]

print(mx_station, round(mx*100, 2))
```
